## Conversão de imagem RGB em imagem Grayscale

```python
import cv2 #importa a biblioteca Cv2
import numpy as np #Importa a biblioteca Numpy para a utilização de matrizes
import matplotlib.pyplot as plt #Importa a biblioteca Matplotlib
#caso for usar o Google Colab com a OpenCV, usar a lib abaixo
from google.colab.patches import cv2_imshow
```

```python
#abrir a imagem
img = cv2.imread('t1.png',1) #lê a imagem
#caso for usar o Google Colab com a OpenCV,
cv2_imshow(img) #exibe a imagem
```

```python
#mostrando a imagem colorida - caso você use Python no seu computador
#reutilize para exibir as imagens em outros códigos
#Voce pode descomentar o código apagando as aspas simples
'''
cv2.imshow('in', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
'''
```

```python
#aplicando conversão ponderada - converte imagem colorida para preto e branco
#img_grayscale_basic = 0.299*img[ : , : ,0] + 0.587*img[ : , : ,1] + 0.114*img[ : , : ,2]

#cv2
B, G, R = cv2.split(img) #separa a imagem em escalas de Azul, verde e vermelho
img_grayscale_pondered = 0.299*B+0.587*G+0.114*R #Define uma escala preto e branco da imagem

img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8) #Define a imagem como sendo uma matriz de elementos do tipo uint8

cv2_imshow(img_grayscale_pondered) #Exibe a imagem com os tons de cinza
```

## Transformações

```python
#negativo

#img_negative[ : , : ,0] = 255 - img[ : , : ,0]
#img_negative[ : , : ,1] = 255 - img[ : , : ,1]
#img_negative[ : , : ,2] = 255 - img[ : , : ,2]

#Mude a variavel colorida para 1 caso queira colorida, o para 0 caso queira em escala de cinza
colorida = 1 #Define se a imagem será colorida ou preto e branco
img_in = cv2.imread('t1.png',colorida) #Le o arquivo da imagem

img_out = 255 - img_in #Define cada pixel como sendo a cor branca menos o pixel original

cv2_imshow(img_in) #Exibe a imagem original
cv2_imshow(img_out) #Exibe o resultado
```

```python
#contraste e brilho
img_in = cv2.imread('t1.png',1) #lê a imagem
#altera os valores tanto de a quanto de b

a = 1 #Define um termo que irá multiplicar todos os pixels da imagem
b = 0.299 #Define um termo que será somado em todos os pixels da imagem

img_out = a*img_in + b #Define a imagem de saida como sendo a imagem multiplicada por a e somada com b

img_out = np.array(img_out, dtype = np.uint8) #Define a imagem como sendo uma matriz com elementos do tipo uint8

cv2_imshow(img_in) #Exibe a imagem original
cv2_imshow(img_out) #Exibe o resultado
```

## Filtro Espacial

```python
#suavização
#você pode ler sobre o conceito aqui: https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html
img_in = cv2.imread('t1.png',0) #Lê a imagem

kernel = np.ones((5,5),np.float32)/25 #define o operador morfológico como sendo uma matriz 5X5 de valores 0.25

img_out_1 = cv2.filter2D(img_in,-1,kernel) #Aplicação filtro de desfoque na imagem


cv2_imshow(img_in) #Exibe a imagem original
cv2_imshow(img_out_1) #Exibe a imagem com desfoque
```
